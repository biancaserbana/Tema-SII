```python
from transformers import CamembertTokenizer, CamembertForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import torch
import pandas as pd

df = pd.read_csv('train.csv')

# impartirea datelor în seturi de antrenament și testare
df['Text'] = df['Text'].astype(str)
df['Text'] = df['Text'].fillna("")
df['Label'] = df['Label'].replace({'true': 1, 'fake': 0, 'biased': 2})  # etichetele in valori numerice
print(df['Label'].dtype)
print(df['Label'].unique())
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df['Text'].tolist(), df['Label'].tolist(), test_size=0.2, random_state=0
)
```

```python
df.head()
```

```python
# initializarea tokenizer
model_name = "camembert-base"
tokenizer = CamembertTokenizer.from_pretrained(model_name)

def tokenize_function(texts):
    if isinstance(texts, list):
        return tokenizer(texts, padding="max_length", truncation=True, max_length=512, return_tensors="pt")
    else:
        return tokenizer([texts], padding="max_length", truncation=True, max_length=512, return_tensors="pt")
```

```python
train_encodings = tokenize_function(train_texts)
test_encodings = tokenize_function(test_texts)
```

```python
import torch

class NewsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}

        label = self.labels[idx]

        item["labels"] = torch.tensor(label, dtype=torch.long)
        return item
```

```python
train_dataset = NewsDataset(train_encodings, train_labels)
test_dataset = NewsDataset(test_encodings, test_labels)
```

```python
# initializarea model
model = CamembertForSequenceClassification.from_pretrained(model_name, num_labels=3)  # 3 clase: true/false/biased

# metoda de evaluare
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = torch.argmax(torch.tensor(logits), dim=1).numpy()
    precision, recall, f1, _ = precision_recall_fscore_support(labels, predictions, average='weighted')
    acc = accuracy_score(labels, predictions)
    return {"accuracy": acc, "f1": f1, "precision": precision, "recall": recall}
```

```python
from transformers import Trainer, TrainingArguments

# antrenare
training_args = TrainingArguments(
    output_dir="/content/drive/MyDrive/News_Fake_Predictor/cammbert",          # folderul de ieșire
    evaluation_strategy="epoch",    # evaluare după fiecare epocă
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"
```

```python
# antrenarea modelului
trainer.train()
```

```python
# evaluarea pe setul de testare
results = trainer.evaluate()
print("Rezultate evaluare:", results)
```

```python
from transformers import CamembertTokenizer, CamembertForSequenceClassification
import torch
import pandas as pd

tokenizer = CamembertTokenizer.from_pretrained("/content/drive/MyDrive/News_Fake_Predictor/cammbert")
model = CamembertForSequenceClassification.from_pretrained("/content/drive/MyDrive/News_Fake_Predictor/cammbert")
```

```python
df_new = pd.read_csv('test.csv')
df_new['Text'] = df_new['Text'].astype(str).fillna("")
```

```python
df.head()
```

```python
# tokenizare texte
def tokenize_function(texts):
    return tokenizer(texts, padding="max_length", truncation=True, max_length=512, return_tensors="pt")

new_encodings = tokenize_function(df_new['Text'].tolist())
```

```python
# dataloader pentru inferență
class NewsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings):  # Corect: folosește __init__
        self.encodings = encodings

    def __len__(self):  # Corect: folosește __len__
        return len(self.encodings["input_ids"])

    def __getitem__(self, idx):  # Corect: folosește __getitem__
        item = {key: val[idx] for key, val in self.encodings.items()}
        return item

new_dataset = NewsDataset(new_encodings)
new_loader = torch.utils.data.DataLoader(new_dataset, batch_size=16)
```

```python
# generare predicții
model.eval()
predictions = []
with torch.no_grad():
    for batch in new_loader:
        inputs = {key: val.to(model.device) for key, val in batch.items()}
        outputs = model(**inputs)
        logits = outputs.logits
        batch_predictions = torch.argmax(logits, dim=1).cpu().numpy()
        predictions.extend(batch_predictions)
```

```python
df_new['Predicted_Label'] = predictions
```

```python
df_new.to_csv('predictions.csv', index=False)
```
